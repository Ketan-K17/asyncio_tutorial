### Tasks, coroutines, futures, and awaitables

## Futures

Simply, a future is a python object which stores a result value. It is a 'promise' of a result that will be available in the future.

```python
# you have to set the result explicitly on a future object for it to be 'done'.

from asyncio import Future

my_future = Future()
print(f'Is my_future done? {my_future.done()}') # false

# set the result of the future
my_future.set_result(42)
print(f'Is my_future done? {my_future.done()}') # true

print(f'What is the result of my_future? {my_future.result()}')
```

Tasks use futures under the hood.

```python
# awaiting a future

from asyncio import Future
import asyncio

def make_request() -> Future:
    future = Future()
    asyncio.create_task(set_future_value(future))
    return future

async def set_future_value(future: Future) -> None:
    await asyncio.sleep(1)
    future.set_result(42)

async def main():
    future = make_request()
    print(f'Is the future done? {future.done()}')
    value = await future
    print(f'Is the future done? {future.done()}')
    print(value)
    
asyncio.run(main())
```

We won't need to interact with futures directly unless we're reading asyncio api code, a lot of which do use futures under the hood... It's required we understand what a future basically is to read that code.

As mentioned, tasks inherit directly from futures. A task is kind of a combination of a future & a coroutine... it embodies a target function in it, while also not blocking until the result of the function is ready..

All 3 can be used in an await expression, they are hence aka 'awaitables'.

The inheritance diagram - 

Awaitable
├── Future
│   └── Task
└── Coroutine

Task wraps a Coroutine

all awaitables implement the __await__ dunder method, which is written as abstract in the actual awaitable definition. If you implement it for a custom awaitable, you can also use that with await..

```python
import asyncio
from util import async_timed

@async_timed()
async def delay(delay_seconds: int) -> int:
    print(f'sleeping for {delay_seconds} second(s)')
    await asyncio.sleep(delay_seconds)
    print(f'finished sleeping for {delay_seconds} second(s)')
    return delay_seconds

@async_timed()
async def main():
    task_one = asyncio.create_task(delay(2))
    task_two = asyncio.create_task(delay(3))
    await task_one
    await task_two
    
asyncio.run(main())
```
