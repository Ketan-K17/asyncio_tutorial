### Event loop

A single-threaded scheduler that runs continuously, advancing coroutines and firing callbacks as their awaited I/O operations or timers become ready — one at a time, but efficiently interleaved.

```python
# a very basic event loop.

from collections import deque
messages = deque()
while True:
    if messages:
        message = messages.pop()
        process_message(message)
```

### How an event loop runs, very roughly

- each iteration of that indefinite while loop can be considered an iteration where the event loop does its job.

- in each iteration, the event loop looks for tasks to do. Once it finds one, it executes it, until it reaches an IO request. At this pt, the task is 'paused' a socket is registered by the event loop for the said task. And this socket is then given to the OS to watch out results for. Following the result, the task is 'woken up', os notifies the event loop that this task can now continue.. it does this for every task.

`Each iteration: drain the ready queue → run each ready task until it awaits → register any new I/O/timer interests → call the OS poller (with a timeout = next timer deadline) → move newly-ready tasks back into the ready queue → repeat.`

### Summary

All in all, you can save up on a lot of time using the concept of the single-threaded event loop, and having IO-bound tasks run concurrently, while only one thread runs actual python bytecode. This is what asyncio streamlines for us.
